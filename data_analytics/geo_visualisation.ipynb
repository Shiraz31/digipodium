```python
import pandas as pd
import numpy as np
import plotly.express as px
import folium as fi
from folium.plugins import MarkerCluster
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=20, skipfooter=2)
df.head()
```

```python
px.choropleth(
    data_frame=df
    locationmode='country names',
    locations=df.index,
    projection='natural earth',
    scope='world',
    color='Total',
    hover_name=df.Continent,
    height=600,
    color_continuous_scale='Rainbow',
)
```

```python
incdf=pd.read_csv('incidents_small.csv')
incdf.head()
```

```python
fi.Map(titles='Stamen Terrain', zoom_start=2)
```

```python
map=fi.Map(location=(37.7749,-122.4194), zoom_start=12)
s = incdf.head(100).copy()
for lng, lat, incident in zip(s.Y, s.X, s.Category):
    fi.Marker([lng, lat], popup=incident).add_to(map)
map
```

```python
map=fi.Map(location=(37.7749,-122.4194), zoom_start=12)
s = incdf.head(100).copy()
for lng, lat, incident in zip(s.Y, s.X, s.Category):
    fi.CircleMarker([lng, lat],radius=5,color='red', fill=True, fill_color='red',fill_opacity=0.6,
     popup=incident).add_to(map)
map
```

```python
map=fi.Map(location=(37.7749,-122.4194), zoom_start=12, titles='Stamen Toner')
marker_cluster = MarkerCluster().add_to(map)
for lat, lng, incident in zip(incdf.Y, incdf.X, incdf.Category):
    fi.Marker([lat, lng], popup=incident).add_to(marker_cluster)
map
```
