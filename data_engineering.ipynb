column related operations


```python
df.rename(mapper={"make":"brand"}, axis=1, inplace=True)
```

```python
df.drop(columns=['unnamed: 0'], inplace=True)
```

```python
df.head()
```

```python
df.replace('alfa-romero', 'alfa-romeo')
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.head()
```

```python
df.dropna(subset=['price'])
```

```python
df.info
```

```python
df.isnull().sum()  #count the no of missing data in each column
```

```python
mnl = df['normalized-losses'].astype('float').mean() #also median and sum can find
```

```python
df['normalized-losses'].replace(np.nan, mnl, inplace=True)
df['bore'].replace(np.nan, df['bore'].astype('float').mean(), inplace=True)
df['stroke'].replace(np.nan, df['bore'].astype('float').mean(), inplace=True)
df['horsepower'].replace(np.nan, df['bore'].astype('float').mean(), inplace=True)
df['peak-rpm'].replace(np.nan, df['bore'].astype('float').mean(), inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("automobile_cleaned.csv", index=False)
```
